# 4 Visulization

# 5 Histogram

```python
plt.figure(figsize=(6,8))
sns.histplot(hr["key"],bins=20,kde=True)
plt.title("Key distribution")
plt.show()
```

# 6 Heatmap

```python
hr1=hr.select_dtypes(include=["number"])
hr1.head()
```

```python
plt.figure(figsize=(10,20))
sns.heatmap(hr1.corr(),cmap="viridis",annot=True,linewidth=0.5)
plt.title("Correletion Matrix ")
plt.show()
```

# 7 Boxplot

```python
plt.figure(figsize=(6,5))
sns.boxplot(x=hr["key"])
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,9))
sns.boxplot(x=hr["artist_count"])
plt.title("artist_count Distribution")
plt.show()
```

# 8 Countplot

```python
plt.figure(figsize=(9,10))
sns.countplot(x=hr["key"],palette="pastel",hue=hr["released_month"])
plt.title("Count of Department")
plt.show()
```

```python
plt.figure(figsize=(9,10))
sns.countplot(x=hr["key"],palette="pastel",hue=hr["released_year"])
plt.title("Count of BusinessTravel")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=hr["valence_%"],palette="pastel",hue=hr["valence_%"])
plt.title("Count of Attrition")
plt.show()
```
